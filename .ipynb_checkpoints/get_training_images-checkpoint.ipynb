```python
import cv2
import os
import sys
import time

try:
    # Escolher quantas imagens quer tirar para treino e qual a categoria. 
    # Será criada uma pasta para as imagens de treino e uma subpasta para cada categoria
    num_of_images = 20
    image_label = 'down'
except:
    print('Error')
    exit(-1)

font = cv2.FONT_HERSHEY_PLAIN
click = False

# Cria um subfolder 'training_img_folder' e cada categoria terá uma pasta dentro deste subfolder
training_img_folder = 'training_images'

# Label_name é o nome da categoria
label_name = os.path.join(training_img_folder, image_label)

# Contador para o numero de imagens feitas
count = image_name = 0

try:
    os.mkdir(training_img_folder)
except FileExistsError:
    pass
try:
    os.mkdir(label_name)
except FileExistsError:
    # Se já existe alguma imagem na pasta, considera para corrigir a numeração do nome
    image_name=len(os.listdir(label_name))

    
# Inicia a captura de vídeo da webcam
video = cv2.VideoCapture(0)
video.set(cv2.CAP_PROP_FRAME_WIDTH, 2000)
video.set(cv2.CAP_PROP_FRAME_HEIGHT, 2000)

while True:
    ret, image = video.read()
    image = cv2.flip(image, 1)
    #time.sleep(0.2)
    if not ret:
        continue

    # Para de gravar quando chega no múmero solicitado
    if count == num_of_images:
        break

    # Desenha um retangulo com borda branca. O que estiver dentro do retangulo será salvo.
    cv2.rectangle(image, (200, 200), (550, 550), (255, 255, 255), 2)

    # Começa a salvar fotos quando apertar a tecla 's'
    if click:
        region_of_interest = image[200:550, 200:550]
        save_path = os.path.join(label_name, '{}.jpg'.format(image_name + 1))
        cv2.imwrite(save_path, region_of_interest)
        image_name += 1
        count += 1


    cv2.putText(image, "Fit the gesture inside the white box and Press 's' key to start clicking pictures",
            (20, 30), font, 0.8, (255, 0, 0), 1, cv2.LINE_AA)
    cv2.putText(image, "Press 'q' to exit.",
            (20, 60), font, 0.8, (255, 0, 0), 1, cv2.LINE_AA)
    cv2.putText(image, "Image Count: {}".format(count),
            (20, 100), font, 1, (12, 20, 200), 2, cv2.LINE_AA)
    cv2.imshow("Get Training Images", image)

    k = cv2.waitKey(10)
    if k==ord('q'):
            break
    if k == ord('s'):
        click = not click

print("\n\nDone\n\n")
video.release()
cv2.destroyAllWindows()
```
