```python
import cv2
import numpy as np
from keras_squeezenet import SqueezeNet
from tensorflow.keras.optimizers import Adam
from keras.utils import np_utils
from keras.layers import Activation, Dropout, Convolution2D, GlobalAveragePooling2D
from keras.models import Sequential
import tensorflow as tf
from tensorflow.python.keras.optimizers import TFOptimizer
import os

# Seta os parametros do medelo de treinamento
def model_parameters():
    length_categories = len(CATEGORY)
    base_model = Sequential()
    base_model.add(SqueezeNet(input_shape=(225, 225, 3), include_top=False))
    base_model.add(Dropout(0.5))
    base_model.add(Convolution2D(length_categories, (1, 1), padding='valid'))
    base_model.add(Activation('relu'))
    base_model.add(GlobalAveragePooling2D())
    base_model.add(Activation('softmax'))

    return base_model

# Retorna os valores numericos das classes
def label_mapper(val):
    return CATEGORY[val]

# Pasta das imagens de treino
training_images = 'training_images'

# Definindo valores das categorias de classificacao
CATEGORY = {
    "up": 0,
    "down": 1
}
    
# Carregando todas as imagens do folder training_images para treinamento
input_data = []
for sub_folder_name in os.listdir(training_images):
    path = os.path.join(training_images, sub_folder_name)
    for fileName in os.listdir(path):
        if fileName.endswith(".jpg"):
            image = cv2.imread(os.path.join(path, fileName))
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            image = cv2.resize(image, (225, 225))
            # Salvando a imagem e os valores no vetor input_data
            input_data.append([image, sub_folder_name])

# Função para separar as imagens de entrada do texto de saida 
image_data, labels = zip(*input_data)

# Converte para numero as classes das imagens
labels = list(map(label_mapper, labels))

# Cria uma lista com os valores das imagens treinadas
labels = np_utils.to_categorical(labels)

# Compilando o modelo
model = model_parameters()
model.compile(
    optimizer=Adam(lr=0.0001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Fitando o modelo
model.fit(np.array(image_data), np.array(labels), epochs=15)

print("Training Completed")

# Salvando o modelo de treino em um arquivo .h5
model.save("treinamento_gestos.h5")
```
